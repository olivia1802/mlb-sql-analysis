# Exploring MLB Offensive Performance Using SQL

This project explores offensive performance trends across MLB players and teams using SQL queries applied to 2023 batting statistics. The analysis focuses on team offensive production, plate discipline, and player performance beyond traditional home run metrics.

```python
import pandas as pd
import sqlite3

df = pd.read_csv("Batting.csv", sep=";", encoding="latin1")

conn = sqlite3.connect(':memory:')
df.to_sql('batting', conn, if_exists='replace', index=False)
```

## Exploratory Analysis

Which players ranked among the league leaders in home runs?

```python
query = """
SELECT Name,
       Tm,
       HR
FROM batting
ORDER BY HR DESC
LIMIT 10
"""

pd.read_sql_query(query, conn)
```

### Findings

Shohei Ohtani, Matt Olson, and Luis Robert Jr. ranked among the league leaders in home runs within this dataset. The results provide an initial overview of the league's top power hitters and establish context for the team-level analyses that follow.

## Research Question 1

Which teams demonstrated the strongest offensive production based on average OPS?

```python
query = """
SELECT Tm,
       ROUND(AVG(OPS),3) AS Avg_OPS
FROM batting
WHERE Tm != 'TOT'
GROUP BY Tm
ORDER BY Avg_OPS DESC;
"""

pd.read_sql_query(query, conn)
```

### Findings

MIL, PHI, and MIA recorded the highest average OPS values in the dataset, indicating stronger offensive production relative to other teams in the league.

## Research Question 2

How do team home run totals compare with average offensive production?

```python
query = """
SELECT Tm,
       SUM(HR) AS Total_HR,
       ROUND(AVG(OPS),3) AS Avg_OPS
FROM batting
WHERE Tm != 'TOT'
GROUP BY Tm
ORDER BY Total_HR DESC;
"""

pd.read_sql_query(query, conn)
```

### Findings

ATL, LAD, and LAA recorded the highest team home run totals in the dataset. However, teams with the most home runs did not always have the highest average OPS. For example, MIL and PHI ranked among the league leaders in average OPS despite recording fewer total home runs than several other teams. This suggests that offensive production is influenced by factors beyond home run hitting alone.

## Research Question 3

Which teams demonstrated the strongest plate discipline based on walk-to-strikeout ratio?

```python
query = """
SELECT Tm,
       ROUND(
           SUM(BB) * 1.0 / SUM(SO),
           3
       ) AS BB_SO_Ratio
FROM batting
WHERE Tm != 'TOT'
GROUP BY Tm
ORDER BY BB_SO_Ratio DESC;
"""

pd.read_sql_query(query, conn)
```

### Findings

SDP and LAD recorded the highest walk-to-strikeout ratios in the dataset, followed by ARI, NYM, and CLE. These teams generated walks at a higher rate relative to strikeouts, suggesting stronger plate discipline and a greater ability to avoid unproductive plate appearances.

## Research Question 4

Which players generated strong offensive production despite relatively low home run totals?

```python
query = """
SELECT Name,
       Tm,
       PA,
       HR,
       ROUND(OPS,3) AS OPS
FROM batting
WHERE OPS >= 0.850
  AND HR < 20
  AND PA >= 100
ORDER BY OPS DESC;
"""

pd.read_sql_query(query, conn)
```

### Findings

Several players maintained strong offensive performance despite recording fewer than 20 home runs. Notably, Luis Arraez posted an OPS of .892 with only three home runs, while Juan Soto and Yandy Díaz recorded OPS values above .890 without relying heavily on home run production. These results suggest that offensive value can be generated through on-base ability, contact hitting, and overall offensive consistency rather than home run power alone.

## Conclusion

While teams such as ATL led the league in home run production, teams including MIL and PHI ranked highest in average OPS. In addition, players such as Luis Arraez and Juan Soto maintained strong offensive performance despite relatively low home run totals.
